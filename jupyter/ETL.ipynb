## Limpieza

### Tratamiento de valores nulos

```python
# Elimina las filas q  contengas por lo menos un valor nulo
df_sinNulos=salaries.dropna()

# Elimina las columnas q contengan por lo menos un valor nulo
df_sinNulos_columnas=df_sinNulos.dropna(axis=1)
```

```python
df_sinNulos_columnas.shape
```

```python
df_sinNulos["salary"][8]
```

Ejecutamos extract para extraer (numeros, comas, puntos) del primer numero, q es el q nos interesa, y despues reemplazamos la "," por "" y convertimos en tipo float

y lo almacenamos en una columna llamada "Salary_numeric"

```python
df_sinNulos["Salary_numeric"]= df_sinNulos['salary'].str.extract(r'\$([\d,]+\.\d+)')[0].str.replace(',', '').astype(float)
df_sinNulos["Salary_numeric"]
```

```python
df_sinNulos["Salary_numeric"].describe()
```

```python
#crear un boxplot
plt.boxplot(df_sinNulos["Salary_numeric"])

#Añador etiquetas y titulos
plt.xlabel("Salarios")
plt.ylabel("valores en USD")
plt.title("Boxplot")

#Mostrar grafico
plt.show()
```

```python
#Crear un histograma
plt.hist(df_sinNulos["Salary_numeric"], bins=30, color="blue", alpha=0.7)

#Añador etiquetas y titulos
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma de datos aleatorios")

#Mostrar grafico
plt.show()
```

como podemos observar que la distribucion de los salarios esta desvalanceada, y esto se debe porque hay puestos de trabajo que muestran el salario anul y otros por hora

### corregir los datos para mostrar solo los que tengas salarios anuales

```python
#eliminar las filas q contengas un salario menor a 20000
df_sinNulos=df_sinNulos[df_sinNulos["Salary_numeric"]>=20000]
```

```python
#crear un boxplot
plt.boxplot(df_sinNulos["Salary_numeric"])

#Añador etiquetas y titulos
plt.xlabel("Salarios")
plt.ylabel("valores en USD")
plt.title("Boxplot")

#Mostrar grafico
plt.show()
```

```python
#Crear un histograma
plt.hist(df_sinNulos["Salary_numeric"], bins=30, color="blue", alpha=0.7)

#Añador etiquetas y titulos
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma de datos aleatorios")

#Mostrar grafico
plt.show()
```

reemplazamos los valores nulos con la media de la columna, aunque los valores nulos ya fueron eliminados de la columna

```python
df_final=df_sinNulos.fillna(df_sinNulos["Salary_numeric"].mean())
df_final
```

### eliminamos la colunma "salary"

```python
df_final=df_final.drop("salary", axis=1)
```

```python
df_final
```

### Agrupamos por title y calculamos el mean , median , count

```python
df_agrupado=df_final.groupby("title")["Salary_numeric"].agg(["mean","median","count"]).reset_index()
df_agrupado
```

### cambiamos el nombre de las columnas para mayor claridad

```python
df_agrupado.columns=["title","salary_mean","salary_median","salary_count"]
df_agrupado
```

```python
df_agrupado.info()
```

```python
# Transformar a datos tipo int
df_agrupado["salary_mean"]=df_agrupado["salary_mean"].astype(int)
df_agrupado["salary_median"]=df_agrupado["salary_median"].astype(int)

df_agrupado
```

### Exportar CSV

```python
# Transformar a archivo CSV y exportar
df_agrupado.to_csv("./archivoFinal/linkedin_jobs_usa.csv")
```
